2. Defina una función de costo y una heurística para su representación. ¿Es su heurística admisible? (Justifique su respuesta)

El costo de un estado a otro es representado como el costo de realizar el movimiento en el 8-puzzle, es decir, 1 movimiento por lo que el valor del costo de un estado al otro es de 1. Para obtener el costo de un estado después de **n** movimientos simplemente se toma el nivel del padre (el nivel representa la cantidad de movimientos realizados hasta ese estado) y se le suma 1, ya que sería el nivel de ese estado actual o hijo que representa la cantidad de movimientos a realizar para llegar a ese estado.

Para la función heurística se utilizó la distancia Manhattan desde cada posición actual hasta su posición correcta [2], por ejemplo:

|&nbsp;|&nbsp;|&nbsp;|
|------|------|------|
| 1 | &nbsp; | 3 |
| 5 | 2 | 6 |
| 4 | 7 | 8 |

Para el caso mostrado la heurística sería:

*   **Para 1:** |1 - 1| + |1 - 1| = 0 + 0 = 0
*   **Para 2:** |1 - 2| + |2 - 2| = 1 + 0 = 1
*   **Para 3:** |1 - 1| + |3 - 3| = 0 + 0 = 0
*   **Para 4:** |2 - 3| + |1 - 1| = 1 + 0 = 1
*   **Para 5:** |2 - 2| + |2 - 1| = 0 + 1 = 1
*   **Para 6:** |2 - 2| + |3 - 3| = 0 + 0 = 0
*   **Para 7:** |3 - 3| + |1 - 2| = 0 + 1 = 1
*   **Para 8:** |3 - 3| + |2 - 3| = 0 + 1 = 1

Cuando se realiza la suma de cada una se tiene que: 

h(n) = 0 + 1 + 0 + 1 + 1 + 0 + 1 + 1 = 5

Tomando en cuenta que [2]:

*  n es un nodo.
*  h es una heurística.
*  h(n) es el la estimación del coste mínimo para alcanzar el objetivo desde n, aplicando h.
*  C(n) es el coste mínimo real para alcanzar el objetivo desde n.

Decimos que h es admisible si para todo n, h(n) <= C(n).

Determinamos que la heurística usada es admisible ya que: C(n) es el coste mínimo para alcanzar desde n al objetivo y la distancia Manhattan calcula la distancia mínima que se puede realizar de un estado al otro, con lo cual siempre se cumplirá para cualquier n en cualquier posición que: h(n) <= C(n). Por ejemplo: para el ejemplo anterior y para el número 2 tenemos que: h(2) = 1 y C(2) = 1, por lo tanto h(2) <= C(2) se cumple.

Escogí esta heurística debido a que busca siempre obtener las distancias de cada número desde su posición actual a su posición original o correcta, por lo que al evaluar f(n) = g(n) + h(n) se procupara escoger el estado en donde sus números estan más cerca a sus posiciones correctas.

```python
def cost(parent_cost):
  return parent_cost + 1

def heuristic(matrix):
     total = 0
     h = 0
     
     for i in range(n):
          for j in range(n):
               if matrix[i][j] == 1:
                    h = abs(0 - i) + abs(0 - j)
               if matrix[i][j] == 2:
                    h = abs(0 - i) + abs(1 - j)
               if matrix[i][j] == 3:
                    h = abs(0 - i) + abs(2 - j)
               if matrix[i][j] == 4:
                    h = abs(1 - i) + abs(0 - j)
               if matrix[i][j] == 5:
                    h = abs(1 - i) + abs(1 - j)
               if matrix[i][j] == 6:
                    h = abs(1 - i) + abs(2 - j)
               if matrix[i][j] == 7:
                    h = abs(2 - i) + abs(0 - j)
               if matrix[i][j] == 8:
                    h = abs(2 - i) + abs(1 - j)
               if matrix[i][j] == 9:
                    h = 0
               total = total + h
     
     return total
```
